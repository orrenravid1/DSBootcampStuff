<h2>Dummy variables</h2>
<img src='https://raw.githubusercontent.com/rucogsciclub/Data_Science_Bootcamp/master/Week_4/resources/dummy.png' width=300></img>
<br>
So we have some categorical variables, but we have no way to manipulate them numerically. Currently, their values are in the form of text. But, what if we could artificially turn their values from text to numbers? 
<br>
Enter <i>dummy variables</i>. What we do is find all the unique values a categorical variable can take on and assign a number arbitrarily to each of the unique values. This allows us to treat our variables <i>as if</i> they are "numeric". <b>But</b>, we will need to be careful about how we use this as we will see later on.

```python
# A function for creating dummy variables
def create_dummies(df, cols):
    new_df = df.copy()
    # For every chosen column, find all the unique values 
    # and collect them in a list
    for col in cols:
        unique = list(df[col].unique())
        # Then, simply assign that value's index in the list as its new value
        new_df[col] = new_df[col].apply(lambda x: unique.index(x))
    return new_df

# Create dummy variables out of all of our categorical features
banking_df = create_dummies(df_raw, categorical_features+['y'])
# Take a look at our new dataframe with dummy variables in it
banking_df.head(10)
```

<h2>Revisiting Multicollinearity</h2>

When we last saw multicollinearity in the linear regression case, all of our variables were numeric. As such, it made sense to look at their relationship in the sense of direct correlations and linear dependence on one another. 
<br>
Now, we have categorical variables. As we saw before, our conversion to dummy variables left us with a non-unique numerical ordering of our values for each categorical value. With this in mind, let's try to examine our correlation matrix and see if it makes sense...

We have wayyyyy too many variables for one correlation matrix!

```python
correlation_matrix = X.corr().round(2)
fig, ax = plt.subplots(figsize=(20,12))
sns.heatmap(data=correlation_matrix, annot=True)
plt.show()
```

Let's split our matrix into 4 equal chunks

```python
# Take the dimesnions of our matrix and split each in half
split_dims = np.array(correlation_matrix.shape, dtype=np.dtype(int))// 2
correlation_split = []
# Take each of these splits and allocate a new plot with one quadrant of
# the original matrix
correlation_split.append(correlation_matrix.iloc[:split_dims[0], :split_dims[1]])
correlation_split.append(correlation_matrix.iloc[:split_dims[0], split_dims[1]:])
correlation_split.append(correlation_matrix.iloc[split_dims[0]:, :split_dims[1]])
correlation_split.append(correlation_matrix.iloc[split_dims[0]:, split_dims[1]:])

# Plot the 4 quadrants from left to right and top to bottom
for cm in correlation_split:
    fig, ax = plt.subplots(figsize=(20,12))
    sns.heatmap(data=cm, annot=True, vmin=0, vmax=1)
    plt.show()
```

<h3>A return to VIF:</h3>
<br>
Now we look at the VIF again. Looking at the conclusions we drew from the correlation matrix, let's see if we run into an analogous problem here...

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor as VIF
```

```python
vif = pd.DataFrame()
vif["VIF Factor"] = [VIF(X.values, i) for i in range(X.shape[1])]
vif["features"] = X.columns
vif
```

A good way to understand why VIF isn't working here is due to a similar reasoning to the correlation matrix. There are two factors that make this not work:

1. Even though we assigned numerical values to our categorical variables, those values and their order is <b>arbitrary</b>! So any ordering of the variables is valid, but each unique ordering might lead to a different VIF.
2. VIF is internally doing multiple linear regression with the features. If linear regression generally doesn't work with categorical variables, then so too must VIF not work with categorical variables.

```python
X_num = X[numerical_features]

vif = pd.DataFrame()
vif["VIF Factor"] = [VIF(X_num.values, i) for i in range(X_num.shape[1])]
vif["features"] = numerical_features
vif
```

<h2>Let's Get Modeling</h2>

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
predictors = X
target = Y

x_train, x_test, y_train, y_test = train_test_split(predictors, target, test_size=0.3, random_state=42)
```

```python
logistic_model = LogisticRegression(solver='lbfgs', max_iter=1e4, C=1e10)
logistic_model.fit(x_train, y_train)
predictions = logistic_model.predict(x_test)
```

```python
accuracy = accuracy_score(y_test, predictions)
print("Accuracy = ", accuracy*100)
```

```python
betas = list(logistic_model.intercept_) + list(logistic_model.coef_[0])
coeff_df = pd.DataFrame()
coeff_df['variable'] = ['constant'] + features
coeff_df['coeff'] = betas
coeff_df
```

```python
fig, ax = plt.subplots(figsize=(20,12))
sns.barplot(x = coeff_df['variable'][1:], y = coeff_df['coeff'][1:])
plt.show()
```

```python
predictions_full = predictions
y_test_full = y_test
```

```python
features_sub = ['education', 'housing', 'age']
predictors = X[feature_sub]
target = Y

x_train, x_test, y_train, y_test = train_test_split(predictors, target, test_size=0.3, random_state=42)
```

```python
logistic_model = LogisticRegression(solver='lbfgs', C=1e10)
logistic_model.fit(x_train, y_train)
predictions = logistic_model.predict(x_test)
```

```python
accuracy = accuracy_score(y_test, predictions)
print("Accuracy = ", accuracy*100)
```

```python
betas = list(logistic_model.intercept_) + list(logistic_model.coef_[0])
coeff_df = pd.DataFrame()
coeff_df['variable'] = ['constant'] + feature_sub
coeff_df['coeff'] = betas
coeff_df
```

```python
fig, ax = plt.subplots(figsize=(20,12))
sns.barplot(x = coeff_df['variable'][1:], y = coeff_df['coeff'][1:])
plt.show()
```

<h2>Confusion Matrix</h2>
<img src='https://raw.githubusercontent.com/rucogsciclub/Data_Science_Bootcamp/master/Week_4/resources/confusion.png' width=300></img>

```python
from sklearn.metrics import confusion_matrix
```

```python
class_names = Y.unique()
class_labels = [str(name) for name in class_names]
class_labels.reverse()

# Confusion matrix for our model with chosen features
cm = confusion_matrix(y_test, predictions, class_names)
cm = np.fliplr(cm)
print(cm)
fig, ax = plt.subplots(figsize=(20,12))
cax = ax.matshow(cm, cmap="coolwarm")
plt.title('Confusion Matrix of Logistic Regression for Chosen Features')
fig.colorbar(cax)
ax.set_xticklabels([''] + class_labels)
ax.set_yticklabels([''] + class_labels)
plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()

# Confusion matrix for our model with all features
cm_full = confusion_matrix(y_test_full, predictions_full)
for row in cm_full:
    row.reverse()
print(cm_full)
fig, ax = plt.subplots(figsize=(20,12))
cax = ax.matshow(cm_full, cmap="coolwarm")
plt.title('Confusion Matrix of Logistic Regression for All Features')
fig.colorbar(cax)
ax.set_xticklabels([''] + class_labels)
ax.set_yticklabels([''] + class_labels)
plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()
```

```python
from sklearn.model_selection import KFold
def logistic_summary(features, regularize=False, num_folds = 10, **kwargs):
    
    predictors = X[features]
    target = Y
    
    kf = kwargs.get('kfold', KFold(n_splits=num_folds))
    
    if regularize:
        inv_reg = 1e10
    else:
        inv_reg = 1
    
    coeff_list = []
    accuracy_list = []
    cm_list = []
    
    for i, (train_index, test_index) in enumerate(kf.split(predictors)):
        print("On fold {} / {}".format(i+1, num_folds))
        
        x_train, x_test = predictors.iloc[train_index], predictors.iloc[test_index]
        y_train, y_test = target.iloc[train_index], target.iloc[test_index]
        
        if len(features) > 4:
            logistic_model = LogisticRegression(solver='lbfgs', C=inv_reg, 
                                                max_iter=1e4)
        else:
            logistic_model = LogisticRegression(solver='lbfgs', C=inv_reg)
        logistic_model.fit(x_train, y_train)
        
        betas = list(logistic_model.intercept_) + list(logistic_model.coef_[0])
        coeff_df = pd.DataFrame()
        coeff_df['variable'] = ['constant'] + features
        coeff_df['coeff'] = betas
        coeff_list.append(coeff_df)
        
        predictions = logistic_model.predict(x_test)
        
        accuracy = accuracy_score(y_test, predictions)
        accuracy_list.append(accuracy)
        
        cm = confusion_matrix(y_test, predictions, class_names)
        cm_list.append(cm)
    
    return kf, coeff_list, accuracy_list, cm_list
```

```python
kf, coeff_list, accuracy_list, cm_list = logistic_summary(features)
```
